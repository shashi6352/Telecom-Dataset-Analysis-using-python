```python
type(A)
```

# Donut chart

```python
plt.pie(A, labels=["Not Churned", "Churned"], colors=["orange", "lime"], 
        startangle=50, shadow=True, radius=2, explode=(0,0.2), autopct="%1.2f%%", pctdistance=0.75)

circle=plt.Circle((0,0),1,color="white")
c=plt.gcf()         #to get current figure

c.gca().add_artist(circle)          #to add the circle in center of pie chart

plt.title("Churned vs Not Churned", pad=100)
plt.show()
#to plot pie chart of data A
```
